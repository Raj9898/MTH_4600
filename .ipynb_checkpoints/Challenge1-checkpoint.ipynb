```python
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def coin_toss(p: float) -> str:
    if rnd.random() <= p: return "heads"
    else: return "tails"
```

```python
# the number of elements in the sequence M 
n = round(1e2)
```

```python
# sample array for M_1, ... , M_n
M = np.random.randint(1,10000,n)
```

```python
M
```

```python
plt.figure(figsize=(15,5))
plt.title('Theoretical Distriubtion for $X$')
plt.ylabel('Probability at some $i$')
plt.xlabel('Value at $i$')
plt.bar(np.arange(1,n+1), M/sum(M), color='brown')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.title('Experimental Distriubtion for $X$')
plt.ylabel('Probability at some $i$')
plt.xlabel('Value at $i$')
plt.bar(np.arange(1,n+1), M/sum(M), color='green')
plt.show()
```

```python
p = 0.5
```

```python
sims = [0]*100
```

```python
for i in range(n):
    rolls = 0
    for _ in range(round(1e6)):
        if (p > M[i]/sum(M)):
            rolls += 1
    sims[i] = rolls/1e6
```

```python
np.array(sims)
```

```python
plt.figure(figsize=(15,5))
plt.title('Experimental Distriubtion for $X$')
plt.ylabel('Probability at some $i$')
plt.xlabel('Value at $i$')
plt.bar(np.arange(1,n+1), np.array(sims), color='green')
plt.show()
```
